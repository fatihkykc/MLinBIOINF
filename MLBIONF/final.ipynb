```python
def editTxt(txt_name):
    txt_name[0] = txt_name[0].replace("\n","").replace(" ","").replace('""','"').split('"')[1:-1]
    
    for i in range(1, len(txt_name)):
        txt_name[i] = txt_name[i].replace("\n","").split('"')[1:-1]

        for j in range(len(txt_name[i])-1, 0, -1):
            x = txt_name[i][j]
            if (x == " "):
                txt_name[i].remove(x)

        txt_name[i][6] = txt_name[i][6][1:-1].split(" ")

        temp = []
        for cc in txt_name[i]:
            if (type(cc) == list):
                for c in cc:
                    temp.append(c)
            else:
                temp.append(cc)

        txt_name[i] = temp
```

```python
editTxt(train)
```
